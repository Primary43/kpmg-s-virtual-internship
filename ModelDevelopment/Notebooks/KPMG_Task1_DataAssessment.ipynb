```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from datetime import date, timedelta
```

```python
raw_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/kpmg/Raws/KPMG_VI_New_raw_data_update_final.xlsx',  sheet_name=['CustomerDemographic', 'CustomerAddress', 'Transactions', 'NewCustomerList'])
```

# 1.0 CustomerDemographic

```python
#load
Customer_df = raw_df.get('CustomerDemographic')
#basic inspect
print('*'*50)
print(Customer_df.info())
print('*'*50)
print(Customer_df.describe())
print('*'*50)
Customer_df = Customer_df.set_index('customer_id')

display(Customer_df.head())
display(Customer_df.tail())

#unnecessary column-------------------------------------------------------
Customer_df.drop('default', axis=1, inplace=True)

#missing value issue--------------------------------------------------
print('*'*100)
print('total missing values: ',Customer_df.isnull().values.sum())
print('total row with missing value: ', Customer_df.isnull().any(axis=1).sum())
print('duplicated value: ', Customer_df.duplicated().sum()) #no duplicated value
print('*'*50)
print(Customer_df.isna().sum())
print('*'*50)
print(Customer_df.isnull().sum()/len(Customer_df)*100)
#visualise missing values
Customer_cols = Customer_df.columns
colours = ['#000099', '#ffff00']
sns.heatmap(Customer_df[Customer_cols].isnull(), cmap=sns.color_palette(colours))
plt.show()
# 6 from 13 columns contain null (1763 in total): last_name 125(3.1%), DOB 87(2.1%), job_title 506(12.6%), job_industry_category 656(16.4%), default 302, tenure 87(2.1%)

#data consistency (cata unique)-----------------------------------------------------
print('*'*100)
print('check unique value of each column')
print('*'*50)
cat_cols = Customer_df.select_dtypes(include=object).columns.tolist()

for col in Customer_df[cat_cols]:
    print(col,':')
    print(Customer_df[col].sort_values().unique())
    print('---------------------')

#gender inconsistent misspelling/ job_industry nan/ decease and owncar N Y
#replace inconsistent format in gender
Customer_df['gender'].replace(['F','Femal'], 'Female', inplace=True)
Customer_df['gender'].replace(['M'], 'Male', inplace=True)
Customer_df['gender'].replace(['U'], 'Unknown', inplace=True)
#replace YN with binary (cata to numeric)
Customer_df['deceased_indicator'].replace(['Y'], 1, inplace=True)
Customer_df['deceased_indicator'].replace(['N'], 0, inplace=True)
Customer_df['owns_car'].replace(['Yes'], 1, inplace=True)
Customer_df['owns_car'].replace(['No'], 0, inplace=True)

for col in Customer_df[cat_cols]:
    print(col,':')
    print(Customer_df[col].sort_values().unique())
    print('---------------------')

#outlier issue----------------------------------------------------
print('*'*50)
print(Customer_df['DOB'].describe())
sns.lineplot(x='DOB', y='customer_id' , data=Customer_df)
#remove outlier year 1840
Customer_df = Customer_df[Customer_df['DOB'] != Customer_df.DOB.min()]
sns.lineplot(x='DOB', y='customer_id' , data=Customer_df)
plt.show()


# drop when missing is small portion-------------------------------
Customer_df.dropna(subset=['last_name','DOB','tenure'], inplace=True)
sns.heatmap(Customer_df[Customer_cols].isnull(), cmap=sns.color_palette(colours))
plt.show()
#change data type
Customer_df['tenure'] = Customer_df['tenure'].astype(int)
Customer_df['deceased_indicator'] = Customer_df['deceased_indicator'].astype(bool)
Customer_df['owns_car'] = Customer_df['owns_car'].astype(bool)
print('*'*50)

#from 4000 to 3791 (94.77%) -5.23%
print('other missing values: ',Customer_df.isnull().values.sum()/len(Customer_df)*100)
#other missing value for cata is 29.33%. we will keep it as nan
Customer_df_sort = Customer_df.sort_index()
Customer_df_sort.info()
display(Customer_df_sort)
```

# 2.0 CustomerAddress

```python
#load
Address_df = raw_df.get('CustomerAddress')
#basic inspect
print('*'*50)
print(Address_df.info())
print('*'*50)
print(Address_df.describe())
print('*'*50)
Address_df = Address_df.set_index('customer_id')
display(Address_df.head())
display(Address_df.tail())
#missing value issue--------------------------------------------------
print('*'*100)
print('total missing values: ',Address_df.isnull().values.sum())
print('total row with missing value: ', Address_df.isnull().any(axis=1).sum())
print('duplicated value: ', Address_df.duplicated().sum())
#6 features has no missing value and duplication
#unrelate to 3 customers (1-4003)
#postcode (2000-4882), property_valuation(1-12)

#unnecessary column-------------------------------------------------------
#Address_df.drop('address', axis=1, inplace=True)
Address_df.drop('country', axis=1, inplace=True)

#data consistency (cata unique)-----------------------------------------------------
print('*'*100)
print('check unique value of each column')
print('*'*50)
cat_cols = Address_df.select_dtypes(include=object).columns.tolist()
for col in Address_df[cat_cols]:
    print(col,':')
    print(Address_df[col].sort_values().unique())
    print('---------------------')
#replace inconsistent format in state
Address_df['state'].replace(['New South Wales'], 'NSW', inplace=True)
Address_df['state'].replace(['Victoria'], 'VIC', inplace=True)

for col in Address_df[cat_cols]:
    print(col,':')
    print(Address_df[col].sort_values().unique())
    print('---------------------')

Address_df
```

```python
#3_merge customer table
Customer_df_merged = pd.merge(Customer_df_sort, Address_df, on='customer_id', how='left')
Customer_df_merged.info()
sns.heatmap(Customer_df_merged.isnull(), cmap=sns.color_palette(colours))
plt.show()
#missing value issue--------------------------------------------------
print('*'*100)
print('total missing values: ',Customer_df_merged.isnull().values.sum())
print('total row with missing value: ', Customer_df_merged.isnull().any(axis=1).sum())
print('*'*50)
# drop when small missing portion of 'address', 'postcode', 'state', 'property_valuation'-------------------------------
Customer_df_merged.dropna(subset=['address', 'postcode', 'state', 'property_valuation'], inplace=True)
# has 3787 row left
Customer_df_merged['property_valuation'] = Customer_df_merged['property_valuation'].astype(int)
Customer_df_merged['postcode'] = Customer_df_merged['postcode'].astype(int)
print(Customer_df_merged.info())
print('*'*50)
Customer_df_merged
```

# 3.0 Transactions

```python
#load
Transactions_df = raw_df.get('Transactions')
#basic inspect
print('*'*50)
print(Transactions_df.info())
print('*'*50)
Transactions_df = Transactions_df.set_index('transaction_id')

Transactions_df.head()
print('*'*100)
#missing value issue--------------------------------------------------
print('total missing values: ',Transactions_df.isnull().values.sum())
print('total row with missing value: ', Transactions_df.isnull().any(axis=1).sum())
print('duplicated value: ', Transactions_df.duplicated().sum()) #no duplicated value
print('*'*50)
print(Transactions_df.isna().sum())
print('*'*50)
print(Transactions_df.isnull().sum()/len(Transactions_df)*100)

#contains 20,000 rows with 13 features
#7 columns contain missing/ total missing values:  1542
#online_order(1.8%), brand, product_line, product_class, product_size, standard_cost, product_first_sold_date (0.98%)
#visualise missing values
Transactions_cols = Transactions_df.columns
colours = ['#000099', '#ffff00']
plt.figure(figsize=(7,5))
sns.heatmap(Transactions_df[Transactions_cols].isnull(), cmap=sns.color_palette(colours))
plt.show()

#data consistency (cata unique)-----------------------------------------------------
#has consistent format
print('*'*100)
print('check unique value of each column')
print('*'*50)
cat_cols = Transactions_df.select_dtypes(include=object).columns.tolist()
for col in Transactions_df[cat_cols]:
    print(col,':')
    print(Transactions_df[col].sort_values().unique())
    print('---------------------')

#convert 5digit to datetime---------------------------------------
Transactions_df['product_first_sold_date'] = pd.to_datetime(Transactions_df['product_first_sold_date'], unit='D', origin=pd.Timestamp('1899-12-30'))
print(Transactions_df.info())

#outlier issue----------------------------------------------------
#no outlier
print('*'*50)
print(Transactions_df['transaction_date'].describe())
sns.lineplot(x='transaction_date', y='transaction_id' , data=Transactions_df)
#transaction from 2017/01/01 - 2017/12/30
print('*'*50)
print(Transactions_df['product_first_sold_date'].describe())
sns.lineplot(x='product_first_sold_date', y='transaction_id' , data=Transactions_df)
plt.show()
#product_first_sold_date from 1991/01/21 - 2016/12/06


#numeric outlier--------------------------------------------------------
#'product_id'(0-100), 'customer_id'(1-5034), 'transaction_date', 'online_order'
#'order_status', 'brand', 'product_line', 'product_class', 'product_size', 'list_price'(12-2091.47), 'standard_cost'(7.2-1759.85),
#'product_first_sold_date' (33259-42710)
Transactions_num = Transactions_df.select_dtypes(include='number')
Transactions_df_feature = pd.melt(Transactions_num, var_name="features")
print(Transactions_df_feature)
plt.figure(figsize=(15,9))
sns.boxplot(x="features", y="value", data=Transactions_df_feature)
plt.show()

# drop when missing is small portion-------------------------------
Transactions_df.dropna(inplace=True)
sns.heatmap(Transactions_df.isnull(), cmap=sns.color_palette(colours))
plt.show()

#from 20,000 to 19445 (97.11%)
#change type and format
Transactions_df['online_order'] = Transactions_df['online_order'].astype(bool)
Transactions_df['standard_cost'] = round(Transactions_df['standard_cost'], 2)

print(Transactions_df.info())
print('*'*50)
display(Transactions_df)
```
